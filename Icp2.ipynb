```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/M3IT/COVID-19_Data/master/Data/COVID_AU_cumulative.csv")
df1 = pd.read_csv("https://raw.githubusercontent.com/M3IT/COVID-19_Data/master/Data/COVID_AU_national.csv")
```

```python
df['date'] = pd.to_datetime(df['date'])
df1['date'] = pd.to_datetime(df1['date'])
```

```python
fig1 = px.line(df1, x="date", y="confirmed", title="COVID-19 Confirmed Cases Over Time in Australia",
               labels={"date": "Date", "confirmed": "Confirmed Cases"}, template="plotly_dark")
fig1.show()
```

```python
latest_date = df['date'].max()
df_latest = df[df['date'] == latest_date]

fig2 = px.bar(df_latest, x="administrative_area_level_1", y="confirmed",
              title=f"COVID-19 Cases by State on {latest_date.date()}",
              labels={"administrative_area_level_1": "State", "confirmed": "Total Confirmed Cases"},
              template="plotly_dark")

fig2.show()
```
